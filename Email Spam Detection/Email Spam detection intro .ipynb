**Dataset Features:**

- **Number of Links:** The count of hyperlinks present in the email.
- **Number of Words:** The total words in the email.
- **Attachment:** A binary variable indicating whether the email has an attachment (1 for yes, 0 for no).
- **Spam Label:** The categorical variable representing whether the email is spam or non-spam (ham).
